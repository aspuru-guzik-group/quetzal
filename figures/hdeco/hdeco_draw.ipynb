```python
from chem import Molecule, parse_xyz
import numpy as np
import py3Dmol
from datasets import MoleculeDataset
import torch
from draw import make_html

idxs = [23, 34, 35, 102, 1234, 11111, 668, 948]

mds = MoleculeDataset("qm9_test")

flips = [
    np.array([-1, -1, 1]),
    np.array([-1, 1, -1]),
    np.array([-1, -1, 1]),
    np.array([-1, 1, -1]),
    np.array([1, -1, -1]),
    np.array([-1, 1, -1]),
    np.array([-1, -1, 1]),
    np.array([-1, 1, -1]),
]

view = py3Dmol.view(width=2800, height=2800, viewergrid=(4, 4))
for col in range(4):
    M = mds.mol(idxs[col])
    M.coords = M.coords * flips[col]
    view = M.show(view=view, viewer=(0, col), zoom=False, color="cyan", scale=0.98)
    view = M.show(view=view, viewer=(1, col), zoom=False, color="cyan", scale=0.98)
    view = M.show(view=view, viewer=(2, col), zoom=False, color="cyan", scale=0.98)
    view = M.show(view=view, viewer=(3, col), zoom=False, color="cyan", scale=0.98)

    path = f"samples/hdeco/olex2/test{idxs[col]}.xyz"
    a, x = parse_xyz(path)
    M = Molecule(a, x)
    view = M.show(view=view, viewer=(0, col), zoom=False)

    path = f"samples/hdeco/obabel_hydride/test{idxs[col]}.xyz"
    a, x = parse_xyz(path)
    M = Molecule(a, x)
    view = M.show(view=view, viewer=(1, col), zoom=False)

    path = f"samples/hdeco/symphony/test{idxs[col]}.xyz"
    a, x = parse_xyz(path)
    M = Molecule(a, x)
    view = M.show(view=view, viewer=(2, col), zoom=False)

    path = f"samples/hdeco/quetzal/test{idxs[col]}.xyz"
    a, x = parse_xyz(path)
    x = x * torch.tensor([-1, 1, -1])
    if col == 0:
        x = x * torch.tensor([1, -1, -1])
    if col == 2:
        x = x * torch.tensor([1, -1, 1])
    if col == 4:
        x = x * torch.tensor([-1, -1, 1])
    if col == 6:
        x = x * torch.tensor([1, -1, -1])
    M = Molecule(a, x)
    view = M.show(view=view, viewer=(3, col), zoom=False)

make_html(view, "figures/hdeco/Hdeco_examples1.html")

view = py3Dmol.view(width=2800, height=2800, viewergrid=(4, 4))
for col in range(4, 8):
    M = mds.mol(idxs[col])
    M.coords = M.coords * flips[col]
    view = M.show(view=view, viewer=(0, col-4), zoom=False, color="cyan", scale=0.98)
    view = M.show(view=view, viewer=(1, col-4), zoom=False, color="cyan", scale=0.98)
    view = M.show(view=view, viewer=(2, col-4), zoom=False, color="cyan", scale=0.98)
    view = M.show(view=view, viewer=(3, col-4), zoom=False, color="cyan", scale=0.98)

    path = f"samples/hdeco/olex2/test{idxs[col]}.xyz"
    a, x = parse_xyz(path)
    M = Molecule(a, x)
    view = M.show(view=view, viewer=(0, col-4), zoom=False)

    path = f"samples/hdeco/obabel_hydride/test{idxs[col]}.xyz"
    a, x = parse_xyz(path)
    M = Molecule(a, x)
    view = M.show(view=view, viewer=(1, col-4), zoom=False)

    path = f"samples/hdeco/symphony/test{idxs[col]}.xyz"
    a, x = parse_xyz(path)
    M = Molecule(a, x)
    view = M.show(view=view, viewer=(2, col-4), zoom=False)

    path = f"samples/hdeco/quetzal/test{idxs[col]}.xyz"
    a, x = parse_xyz(path)
    x = x * torch.tensor([-1, 1, -1])
    if col == 0:
        x = x * torch.tensor([1, -1, -1])
    if col == 2:
        x = x * torch.tensor([1, -1, 1])
    if col == 4:
        x = x * torch.tensor([-1, -1, 1])
    if col == 6:
        x = x * torch.tensor([1, -1, -1])
    M = Molecule(a, x)
    view = M.show(view=view, viewer=(3, col-4), zoom=False)

make_html(view, "figures/hdeco/Hdeco_examples2.html")
```
